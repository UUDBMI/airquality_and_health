```python
import pandas as pd
pd.options.display.html.table_schema = True
import numpy as np
```

```python
pm25 = pd.read_csv("Class_PM25_Data.csv")
pm25 = pm25.dropna()
pm25
```

```python
# > 90,000
# 90,000 >= x > 60,000
# 60,000 >= x > 30,000
# 30,000 >= x

grades = []

# For each row in the column,
for row in pm25['Income']:
    # if more than a value,
    if row > 90000 :
        # Append a letter grade
        grades.append('> 90,000')
    # else, if more than a value,
    elif 90000 >= row > 60000:
        # Append a letter grade
        grades.append('90000 >= and < 60000')
    # else, if more than a value,
    elif 60000 >= row > 30000:
        # Append a letter grade
        grades.append('60,000 >= and < 30,000')
    # else, if more than a value,
    elif row < 30000:
        grades.append('< 30,000')
        
# Create a column from the list
pm25['Income_Range'] = grades
pm25
```
